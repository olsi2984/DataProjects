```python
import pandas as pd
```

```python
Prices_R = pd.read_excel("Prices_R.xlsx")
Prices_K = pd.read_excel("Prices_K.xlsx")
```

```python
Prices_R_short = Prices_R[["Kodi", "Përshkrimi", "Kodi 2", "SHUMICA ( EUR )"]].set_index("Kodi")
```

```python
Prices_K_short = Prices_K[["Kodi", "NIV.1 ME TVSH EUR ( EUR )"]].set_index("Kodi")
```

```python
Prices_R_K = pd.concat([Prices_R_short, Prices_K_short], axis=1, sort=False).dropna()
```

```python
Prices_R_K["Dif"] = Prices_R_K["SHUMICA ( EUR )"] - Prices_R_K["NIV.1 ME TVSH EUR ( EUR )"]
```

```python
Prices_R_K_filtered = Prices_R_K.query('Dif != 0 and Përshkrimi != 0').sort_index()
Prices_R_K_filtered.head(1)
```

```python
Prices_R_K_filtered.to_excel("Prices_R_K_filtered.xlsx")
```
